```python
import pandas as pd
```

# JOIN

```python
var1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3']}, index=[0, 1, 2, 3])
var1
var2 = pd.DataFrame({
    'D': ['A4', 'A5', 'A6'],
    'E': ['B4', 'B5', 'B6'],
    'C': ['C4', 'C5', 'C6']}, index=[0, 1, 2])
var2 
```

```python
var1.join(var2, how='outer', lsuffix="_12") # JOIN using left method it will keep all the rows from left dataframe, inner method will keep only the common rows
```

# PIVOT TABLE & MELT FUNCTION
### MELT

```python
var3 = pd.DataFrame({
    "days":[1,2,3,4,5,6],
    "eng":[30,40,50,20,10,60],
    "maths":[35,45,55,25,15,65],
    "science":[32,42,52,22,12,62]

    })

var3
```

```python
pd.melt(var3, id_vars=["days"], var_name="subject", value_name="marks")
```

## PIVOT()

```python
var3 = pd.DataFrame({
    "days":[1,2,3,4,5,6],
    "st_name": ['a','b','c','d','e','b'],
    "eng":[30,40,50,20,10,60],
    "maths":[35,45,55,25,15,65],
    "science":[32,42,52,22,12,62]
})

var3
```

```python
var3.pivot(index="days", columns="st_name",values="eng")
```

```python
var3.pivot(index="days", columns="st_name")
```
