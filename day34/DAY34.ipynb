```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], #paradise biryani
    [2, 120000, 140000, 160000, 190000], #beijing bites
    [3, 200000, 230000, 260000, 300000], #pizza hub
    [4, 180000, 200000, 220000, 250000], # burger point
    [5, 160000, 180000, 200000, 220000] #chai point
])

print("====== Zomato Sales Data ======")
print(sales_data)

```

```python
print("shape:",sales_data.shape)
```

```python
print("sales_data for 1st 3 days :\n", sales_data[:3])
```

```python
#total sales per year
print(np.sum(sales_data, axis=0))
#yearly total
print(np.sum(sales_data[:,1:], axis=0))
```

```python
#minimum sales per restaurent
print(np.min(sales_data[:,1:], axis=1))
#maximum sales per year
print(np.max(sales_data[:,1:], axis=0))
```

```python
#average sales per restaurant
avg_sales_avg = np.mean(sales_data[:,1:], axis=1)
print(avg_sales_avg)
#cumulative
cumsum = np.cumsum(sales_data[:,1:], axis=1)
print(cumsum)
```

```python
plt.figure(figsize=(8,6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average cumulative sales across all restaurant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
