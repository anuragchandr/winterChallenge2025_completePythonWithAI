# Matrix Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('flights')
df
```

```python
df2 = sns.load_dataset('tips')
df2
```

## HeatMap

- correlation

```python
tipscorr = df2[['total_bill', 'tip', 'size']]
tipscorr
```

```python
tipscorr.corr()
```

```python
sns.heatmap(tipscorr.corr(), annot=True, cmap='coolwarm')
```

- clustermap

```python
sns.clustermap(tipscorr.corr(), annot=True, cmap='coolwarm')
```

- Pivot table Heat map

```python
pivotflights = df.pivot(index='month', columns='year', values='passengers')
pivotflights
```

```python
sns.heatmap(pivotflights)
```

# Regression Plots

- implot()

```python
sns.lmplot(x='total_bill', y='tip', data=df2)
```

```python
sns.lmplot(x='total_bill', y='tip',hue='sex', data=df2, markers=['o','v'],scatter_kws={'s':100}, palette='Set1')
```
