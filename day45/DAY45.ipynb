```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,26)
y = x**3
z = x**2
```

```python
plt.plot(x,y)
plt.title("Graph of y = x^2 vs x")
plt.xlabel("X")
plt.ylabel("y")
```

## Subplots

```python

plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(x,z)
```

### Object oriented Way

```python
fig = plt.figure(figsize=(18,12), dpi=300)
axis2 = fig.add_axes([0.1,0.1,0.9,1])
axis2.plot(x,y)
axis1 = fig.add_axes([0.2,0.3,0.3,0.4])
axis1.plot(x,y)
```
