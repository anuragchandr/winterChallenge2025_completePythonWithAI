```python
import pandas as pd
```

# Merging DataFrames

```python
var1 = pd.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})
var2 = pd.DataFrame({"A": [1,2,5], "C": [10, 11, 12]}) #atleast 2 dataframes with same columns
```

```python
merged_df = pd.merge(var1,var2,on ="A") #merging dataframes on column A
merged_df #jo common data hoga uska hi df banega
```

```python
m2df = pd.merge(var1,var2,how ="right") #right join will take all values from right dataframe and matching values from left dataframe
m2df
```

```python
m3df = pd.merge(var1,var2,how ="outer") #outer join will take all values from both dataframes and fill NaN for missing values
m3df
```

```python
var3 = pd.DataFrame({"A": [1, 2, 3], "B": [4, 5, 6]})
var4 = pd.DataFrame({"A": [1,2,5], "B": [10, 11, 12]})
```

```python
m4df = pd.merge(var3,var4,left_index=True,right_index=True, suffixes=("name","id")) #inner join on multiple columns
m4df
```

# Concat

```python
var5 = pd.Series([1,2,3,4])
var6 = pd.Series([5,6,7,8])
```

```python
c1sr = pd.concat([var5,var6],axis=0) #concatinating series 
c1sr
```

```python
d1 = pd.DataFrame({"A": [1,2,3,4], "B": [11,12,13,14]})
d2 = pd.DataFrame({"A": [5,6], "": [7,8]})
```

```python
c2df = pd.concat([d1,d2],axis=1,join="outer") #concatinating dataframes -- inner, outer, left, right
c2df
```

# Groupby

```python
var = pd.DataFrame({"section": ["A", "B", "C", "A", "B", "C", "A", "B", "C"],
                    "name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ian"],
                    "s1": [85, 90, 78, 92, 88, 76, 95, 89, 84],
                    "s2": [80, 85, 82, 88, 90, 79, 91, 87, 83]}) 
var
```

```python
grouped = var.groupby("section") #groupby department and calculate mean salary
grouped
```

```python
for section, group in grouped:
    print(f"Group Name: {section}")
    print(group,"\n")
```
