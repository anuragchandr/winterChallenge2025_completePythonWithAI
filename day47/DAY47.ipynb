# Categorical Plot

- boxplot
- violinplot
- stripplot
- swarmplot
- barplot
- countplot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = sns.load_dataset('tips')
df
```

```python
sns.countplot(x=df['sex'], hue=df['smoker'])
```

```python
sns.barplot(x=df['sex'], y=df['total_bill'])
```

```python
sns.barplot(x=df['sex'], y=df['tip'], estimator=np.sum)
```

```python
# 5 Number parody : 
sns.boxplot(x = df['tip'], y = df['day'],data = df , palette = 'rainbow')
```

```python
sns.violinplot(x=df['day'], y=df['total_bill'], data=df)
```

# Stripplot and swarmplot


```python
sns.stripplot(x=df['day'], y=df['total_bill'], data=df)
```

```python
sns.swarmplot(x=df['day'], y=df['total_bill'], data=df, color='pink', size=5, edgecolor='gray', linewidth=1, alpha=0.7, dodge=True, marker='D', palette='Set2', orient='v', ax=None, zorder=2)
```

```python
sns.violinplot(x=df['day'], y=df['total_bill'], data=df)
sns.swarmplot(x=df['day'], y=df['total_bill'], data=df)
```
