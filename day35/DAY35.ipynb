# Vector Ops

```python
import numpy as np
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])
```

```python
print("Addition:", vector1+vector2)
print("Subtraction:", vector1-vector2)
print("Multiplication:", vector1*vector2)
print("Dot Product:", np.dot(vector1,vector2) )
print("Division:", vector1/vector2)
```

```python
angle = np.arccos(np.dot(vector1,vector2)/(np.linalg.norm(vector1)*np.linalg.norm(vector2)))
print("Angle between vectors:", np.degrees(angle))
```

```python
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], #paradise biryani
    [2, 120000, 140000, 160000, 190000], #beijing bites
    [3, 200000, 230000, 260000, 300000], #pizza hub
    [4, 180000, 200000, 220000, 250000], # burger point
    [5, 160000, 180000, 200000, 220000] #chai point
])

print("====== Zomato Sales Data ======")
print(sales_data)

```

```python
restaurant_types = np.array(['biryani', 'beijing', 'pizza', 'burger', 'chai'])
vectorized_upper = np.vectorize(str.upper)
print(vectorized_upper(restaurant_types))
```

```python
# Broadcast

monthly_avg = sales_data[:,1:] / 12
print(monthly_avg)
```
