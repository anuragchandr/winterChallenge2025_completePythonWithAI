# Arthimetic operation in Python Pandas

```python
var6 = pd.DataFrame({"A":[1,2,3,4], "B":[5,6,7,8]})
print(var6)
```

```python
var6["c"] = var6["A"]*var6["B"] #all ops possible
print(var6)
```

```python
var7 = pd.DataFrame({"A":[10,20,30,40],"B":[15,16,17,18]})
var7
```

```python
var7["python"]= var7["A"]<=20
var7["python2"]= var7["A"]>=16
var7
```

## Delete and modify

```python
var8.insert(1,"python",var8["A"])
var8
```

```python
var8.insert(1,"python2",[11,12,13,14,14])
var8
```

```python
var8["python3"]=var8["A"][:3]
var8
```
