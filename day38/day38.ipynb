### 1. series

````python
data = [10, 20, 30, 40, 50]
x = pd.Series(data)
print(x)
print(type(x))
print(x.index)
print(x.values)
print(x.dtype)
print(x[2])
````

````python

y = pd.Series(data, index=['a', 'b', 'c', 'd', 'e'],dtype=float, name='My Series')
print(y)
````

````python
#from dictionary
data_dict = {"name":['python','c','java','cpp'],"por":[12,13,14,15], "Rank":[1,2,3,4]}
z = pd.Series(data_dict)
print(z)
print(z.dtype)
````

````python
s = pd.Series(12, index=[1,2,3,4,5,6,7])
t = pd.Series(3, index=[1,2,3,4])
print(s)
print(s.dtype)
print(s+t)
````
